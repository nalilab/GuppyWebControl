```python
fluo = im_crop.sum()
print(fluo)
```

make it a simple to use function

```python
def read_fluo(fname):
    img = imageio.imread(fname);
    ymin = 1;ymax=570;
    xmin = 1;xmax = 750;
    im_crop = img[ymin:ymax,xmin:xmax];
    return im_crop.sum()
```
